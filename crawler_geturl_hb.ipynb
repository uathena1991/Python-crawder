```python
from selenium.webdriver.support.ui import WebDriverWait
def find_url(driver):
    element = driver.find_elements_by_tag_name("a")
    if element:
        return element
    else:
        return False
    
def find_click_next(driver):
    element = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/div[1]/div[2]/div[4]/div[2]/ul/li[9]/a")
    if element:
        return element
    else:
        return False

def find_click_previous(driver):
    element = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/div[1]/div[2]/div[4]/div[2]/ul/li[1]/a")
    if element:
        return element
    else:
        return False
    
def find_attribute(driver):
    element = driver.find_elements_by_tag_name("a")
    if element:
        return element
    else:
        return False
# element = WebDriverWait(driver, secs).until(find)
```

```python
outF = open("hb_url_pool_imac_2.txt", "a")
driver = webdriver.Chrome(executable_path = '/Users/uathena/Downloads/chromedriver')  
driver.maximize_window()  
driver.implicitly_wait(4)  
driver.get("http://www.std.gov.cn/hb/hbQuery")
# pdb.set_trace()
wait_max_secs = 500 # s
print (driver.current_url)  # current_url 方法可以得到当前页面的URL  
# print driver.title
max_page = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/div[1]/div[2]/div[4]/div[2]/ul/li[8]/a")
print int(max_page.get_property("text"))
# pdb.set_trace()
page_id = 0
```

```python
## get all urls from http://www.std.gov.cn/db/dbQuery
## may break... if break, continue from current page
# open a file

# for page_id in range(67, int(max_page.get_property("text"))):
while page_id  <= 3992:
    page_id += 1
    #  find current useful urls
#     driver.implicitly_wait(4)  # 2s will crash sometimes, 3s is good, maybe can try 2.5 s??
    time.sleep(1.5)
    if page_id % 100 == 0:
        time.sleep(10)
#     pdb.set_trace()
    tmp_links = []
    tmp_links = WebDriverWait(driver, wait_max_secs).until(find_url)
#     print tmp_links
    tmp_i = -1
    tmp_count = 0
    for link in tmp_links:
        tmp_url = []
        tmp_i += 1
        try:
            tmp_url = link.get_attribute("href")
        except StaleElementReferenceException:
            if tmp_count == 15:
                break
            print "Warning: no url found at %d_%d" %(page_id, tmp_i)
#             pdb.set_trace()
            failed_pool.append(tmp_links)
            # go back to previous page
            click_pre_btn = WebDriverWait(driver, wait_max_secs).until(find_click_previous)
            ActionChains(driver).click(click_pre_btn).perform()  # 链式用法
            page_id -= 1
            print 'Success: go back to previous page'
            break
#          pdb.set_trace()
        if "stdHBDetailed?id=" in tmp_url:
            tmp_count += 1
            url_pool.append(tmp_url)
            print >>outF, tmp_url
    print page_id, 'Valid links', tmp_count 
                
    ## click, next page
    click_next_btn = WebDriverWait(driver, wait_max_secs).until(find_click_next)
    ActionChains(driver).click(click_next_btn).perform()  # 链式用法
```

```python
outF.close()
driver.close()
driver.quit()  
```
