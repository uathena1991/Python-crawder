```python
from selenium.webdriver.support.ui import WebDriverWait
def find_url(driver):
    element = driver.find_elements_by_tag_name("a")
    if element:
        return element
    else:
        return False
def find_attribute(driver):
    element = driver.find_elements_by_tag_name("a")
    if element:
        return element
    else:
        return False
# element = WebDriverWait(driver, secs).until(find)
```

```python
## get all urls from http://www.std.gov.cn/db/dbQuery
## may break... if break, continue from current page
# open a file
outF = open("hb_url_pool.txt", "a")
driver = webdriver.Chrome(executable_path = '/Users/uathena/Downloads/chromedriver')  
driver.maximize_window()  
driver.implicitly_wait(4)  
driver.get("http://www.std.gov.cn/hb/hbQuery")
wait_max_secs = 500 # s
print (driver.current_url)  # current_url 方法可以得到当前页面的URL  
# print driver.title
max_page = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/div[1]/div[2]/div[4]/div[2]/ul/li[8]/a")
print int(max_page.get_property("text"))
# pdb.set_trace()
for page_id in range(int(max_page.get_property("text"))):
    #  find current useful urls
#     driver.implicitly_wait(4)  # 2s will crash sometimes, 3s is good, maybe can try 2.5 s??
    time.sleep(1)
    print page_id
    if page_id % 100 == 0:
        time.sleep(10)
#     pdb.set_trace()
    tmp_links = WebDriverWait(driver, wait_max_secs).until(find_url)
#     print tmp_links
    for link in tmp_links:
        try:
            tmp_rul = link.get_attribute("href")
        except ValueError:
            pdb.set_trace()
        if "stdHBDetailed?id=" in tmp_rul:
            url_pool.append(tmp_rul)
            print >>outF, tmp_rul

    ## next page
    click_btn = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/div[1]/div[2]/div[4]/div[2]/ul/li[9]/a")  # 单击按钮
    ActionChains(driver).click(click_btn).perform()  # 链式用法
outF.close()
driver.close()
driver.quit()  
```
