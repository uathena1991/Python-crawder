```python
## get all urls from http://www.std.gov.cn/db/dbQuery
## may break... if break, continue from current page
url_pool = [] 


driver = webdriver.Chrome()  
driver.maximize_window()  
driver.implicitly_wait(4)  
  
driver.get("http://www.std.gov.cn/db/dbQuery")
print (driver.current_url)  # current_url 方法可以得到当前页面的URL  
print driver.title
max_page = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/div[1]/div[2]/div[4]/div[2]/ul/li[8]/a")
for page_id in range(int(max_page.get_property("text"))):
    #  find current useful urls 
    print page_id
    time.sleep(3)  # 2s will crash sometimes, 3s is good, maybe can try 2.5 s??
#     pdb.set_trace()
    for link in driver.find_elements_by_tag_name("a"):
        try:
            tmp_rul = link.get_attribute("href")
        except ValueError:
            pdb.set_trace()
        if "stdDBDetailed?id=" in tmp_rul:
            url_pool.append(tmp_rul)
    ## next page
    click_btn = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/div[1]/div[2]/div[4]/div[2]/ul/li[9]/a")  # 单击按钮
    ActionChains(driver).click(click_btn).perform()  # 链式用法

driver.close()
driver.quit()  


```

```python
# for each page, find all the urls
idx = 0
for url in url_pool:
    # find all the urls    
    idx += 1
    while True:
        time.sleep(1)
#         SCSF_id = str(case_id).zfill(8)
        SCSF_id = str(1)
        if idx %100 == 0:
            print idx
        # for debug
#         SCSF_id = str(5117)
        # print cnki_url
        try:
            response = urllib2.urlopen(url)  
        except urllib2.URLError, e:
            print str(e.reason)
            continue
        break
    html = response.read()  
#     print html
#     pdb.set_trace()
    #？？？？？
    
    if "未找到该条记录" in html:
        continue
    sp = html.split("\n")
    # SCSF_id = "SCSF00000170"

    # print sp
    for i, info in enumerate(sp):
        if "TC号/SC号" in info:
#             pdb.set_trace()
            tag_sp = sp[i + 1].split('<td align="left" >')
            tc_number = tag_sp[-1].strip("</td>\r")
#             print tc_number
        if "TC/SWG名称" in info:
            tag_sp = sp[i + 1].split('<td align="left" >')
            tc_name = tag_sp[-1].strip("</td>\r")
#             print tc_name

        if "负责专业范围" in info:
            tag_sp = sp[i + 1].split('<td align="left" >')
            prof_range =  tag_sp[-1].strip("</td>\r")
#             print prof_range

        if "秘书长" in info:
            tag_sp = sp[i + 1].split('<td align="left" >')
            mishuzhang = tag_sp[-1].strip("</td>\r")
#             print mishuzhang

        if "秘书处所在单位" in info:
            tag_sp = sp[i + 1].split('<td align="left" >')
            mishuzhang_dept = tag_sp[-1].strip("</td>\r")
#             print mishuzhang_dept

        if "单位地址" in info:
            tag_sp = sp[i + 1].split('<td align="left" >')
            addr =  tag_sp[-1].strip("</td>\r")
#             print "address", addr

        if "联系人" in info:
            tag_sp = sp[i + 1].split('<td align="left" >')
            contact =  tag_sp[-1].strip("</td>\r")
#             print "contact", contact

        if "邮政编码" in info and "bottom content" not in info:
#             pdb.set_trace()
            tag_sp = sp[i + 1].split('<td align="left" >')
            zip_code = tag_sp[-1].strip("</td>\r")
#             print "zip_code", zip_code


        if "电话" in info:
            # print "haha"
            tag_sp = sp[i + 1].split('<td align="left" >')
            tel = tag_sp[-1].strip("</td>\r")
#             print "tel", tel

        if "传真" in info:
            tag_sp = sp[i + 1].split('<td align="left" >')
            fax = tag_sp[-1].strip("</td>\r")
#             print "fax", fax

        if "邮箱" in info and "邮箱登录" not in info:
            tag_sp = sp[i + 1].split('<td align="left" >')
            email = tag_sp[-1].strip("</td>\r")
#             print "email", email,'\n'
    
    output_tbl = [SCSF_id, tc_number, tc_name, prof_range, mishuzhang, mishuzhang_dept, addr, contact, zip_code, tel, fax, email]
#     pdb.set_trace()
    if len(tc_number) > 0:
        # encode unit-8 (chinese)
        [table.write(row_idx, i, output_tbl[i]) for i in range(len(output_tbl))]
        row_idx += 1
        tc_number = []
#         print >> output, "\t".join(output_tbl)

output.save("/Volumes/data1/yefei_local.xls")
```
