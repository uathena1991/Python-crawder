```python
# combine data from several excel files
df = pd.DataFrame()
path = '/Volumes/data1/'
files_xls = ['yefei_0_3992.xls', 'yefei_3392_4891.xls', 'yefei_4891_5000.xls',
            'yefei_5000_7000.xls', 'yefei_7000_10000.xls', 'yefei_10000_13892.xls',
            'yefei_13892_15000.xls', 'yefei_15000_17000.xls', 'yefei_17000_20000.xls',
            'yefei_20000_last.xls']
for f in files_xls:
    data = pd.read_excel(path+f, header=None)
    df = df.append(data)
```

```python
# add headers
df.columns  = ['名称', '标准号', '发布日期', '实施日期', '废止时间', '中国标准分类号', '国际标准分类号',
               '主管部门', '行业分类', '起草单位','起草人', '备案号', '备案公告']
# switch two columns
df = df[['名称', '标准号', '发布日期', '实施日期', '废止时间', '中国标准分类号', '国际标准分类号',
               '主管部门', '行业分类','起草人', '备案号', '备案公告', '起草单位']]
```
